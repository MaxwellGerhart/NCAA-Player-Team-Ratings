```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup


from plottable import ColumnDefinition, Table
from plottable.cmap import normed_cmap
from plottable.plots import image
from plottable.plots import *
```

```python
df = pd.read_csv(r'C:\Users\maxwe\Player and Team Ratings\NCAA\ncaa_ratings.csv')
```

```python
df = df.drop(columns='Unnamed: 0')
```

```python
df = df.sort_values(by='MAX', ascending=False)
```

```python
data = df.copy()
```

```python
df = data.copy()
df = df.sort_values(by='MAX', ascending=False).reset_index(drop=True)
df['Rank'] = df.index + 1
df['MAX'] = df['MAX'].astype(int)
```

```python
df = data.copy()

df['Logo'] = (r'C:\Users\maxwe\America Scouted\Plus-Minus\Logos_png\\' + df['Team'].astype(str) + '.png')
df['ATT'] = df['ATT'].round(2)
df['DEF'] = df['DEF'].round(2)
df = df.sort_values(by='MAX', ascending=False).reset_index(drop=True)
df['Rank'] = df.index + 1
df['MAX'] = df['MAX'].astype(int)


col_defs = [
    ColumnDefinition(
        name="Rank",
        title="",
        textprops={"ha": "left", "weight": "bold"},
        width=0.25,
    ),
    ColumnDefinition(
        name="Logo",
        title="",
        textprops={"ha": "left", "weight": "bold"},
        width=0.5,
        plot_fn=image,
    ),
    ColumnDefinition(
        name="Team",
        title="Team",
        textprops={"ha": "left", "weight": "bold"},
        width=1,
    ),
    ColumnDefinition(
        name="Conference",
        title="Conference",
        textprops={"ha": "left", "weight": "bold"},
        width=1.5,
    ),
    ColumnDefinition(
        name="ATT",
        title='ATT',
        textprops={"ha": "center"},
        cmap=normed_cmap(df["DEF"], cmap=matplotlib.cm.PiYG, num_stds=1),
        width=0.5
    ),
    ColumnDefinition(
        name="DEF",
        title='DEF',
        textprops={"ha": "center"},
        cmap=normed_cmap(df["DEF"], cmap=matplotlib.cm.PiYG_r, num_stds=1),
        width=0.5
    ),
    ColumnDefinition(
        name="MAX",
        title="MAX",
        textprops={"ha": "center", "weight": "bold"},
        width=0.5,
    )
]

df = df.reset_index(drop=True)
df = df.set_index("Rank")
df = df.head(10)
df = df[['Logo', 'Team', 'Conference', 'ATT', 'DEF', 'MAX']]
```

```python
from matplotlib import font_manager as fm
font_properties = fm.FontProperties(fname=r'C:\Users\maxwe\anaconda3\pkgs\r-rmarkdown-1.12-r36h6115d3f_0\lib\R\library\rmarkdown\rmd\h\bootstrap\css\fonts\RobotoMedium.ttf', size=15)
text_color = 'Black'
```

```python
fig, ax = plt.subplots(figsize=(12, 12))


table = Table(
    df,
    column_definitions=col_defs,
    row_dividers=True,
    footer_divider=True,
    ax=ax,
    textprops={"fontsize": 18, "ha": "center", "fontproperties": font_properties, 'color':text_color},
    row_divider_kw={"linewidth": 1, "linestyle": (0, (1, 5))},
    col_label_divider_kw={"linewidth": 1, "linestyle": "-"},
    column_border_kw={"linewidth": 1, "linestyle": "-"},
)



plt.show()
```
