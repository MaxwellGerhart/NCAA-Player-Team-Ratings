```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup


from plottable import ColumnDefinition, Table
from plottable.cmap import normed_cmap
from plottable.plots import image
from plottable.plots import *
```

```python
df = pd.read_csv('PlayerRatings.csv')
```

```python
df = df.drop(columns='Unnamed: 0')
```

```python
data = df.copy()
```

```python
data.head(20)
```

```python
df = data.copy()
df = df[df['Position']=='Defender']
df['Logo'] = (r'C:\Users\maxwe\Statlink\Logos_PNG/' + df['Team'].astype(str) + '.png')
df = df.head(15)

col_defs = [
    ColumnDefinition(
        name="Logo",
        title="Team",
        textprops={"ha": "left", "weight": "bold"},
        width=0.5,
        plot_fn=image,
    ),
    ColumnDefinition(
        name="Team",
        title="",
        textprops={"ha": "center", "weight": "bold"},
        width=1.5,
    ),
    ColumnDefinition(
        name="Name",
        title="Name",
        textprops={"ha": "center", "weight": "bold"},
        width=1.5,
    ),
    ColumnDefinition(
        name="Minutes Played",
        title="",
        textprops={"ha": "center", "weight": "bold"},
        width=0.5,
    ),
    ColumnDefinition(
        name="Goals",
        title="Goals",
        textprops={"ha": "center", "weight": "bold"},
        width=0.5,
    ),
    ColumnDefinition(
        name="Assists",
        title="Assists",
        textprops={"ha": "center", "weight": "bold"},
        width=0.5,
    ),
    ColumnDefinition(
        name="Shots",
        title="Shots",
        textprops={"ha": "center", "weight": "bold"},
        width=0.5,
    ),
    ColumnDefinition(
        name="MAX",
        title='MAX',
        textprops={"ha": "center"},
        cmap=normed_cmap(df["MAX"], cmap=matplotlib.cm.Blues, num_stds=1),
        width=0.5
    )
]

df = df.head(10)
df = df.reset_index(drop=True)
df = df.set_index("Logo")
df = df[['Team', 'Name', 'Minutes Played', 'Goals', 'Assists', 'Shots', 'MAX']]
df['Name'] = df['Name'].apply(lambda x: x.split()[0] + ' ' + x.split()[-1].title())
df = df[['Name', 'MAX']]
```

```python
from matplotlib import font_manager as fm
font_properties = fm.FontProperties(fname=r'C:\Users\maxwe\anaconda3\pkgs\r-rmarkdown-1.12-r36h6115d3f_0\lib\R\library\rmarkdown\rmd\h\bootstrap\css\fonts\RobotoMedium.ttf', size=15)
text_color = 'Black'
```

```python
fig, ax = plt.subplots(figsize=(6, 12))


table = Table(
    df,
    column_definitions=col_defs,
    row_dividers=True,
    footer_divider=True,
    ax=ax,
    textprops={"fontsize": 16, "ha": "center", "fontproperties": font_properties, 'color':text_color},
    row_divider_kw={"linewidth": 1, "linestyle": (0, (1, 5))},
    col_label_divider_kw={"linewidth": 1, "linestyle": "-"},
    column_border_kw={"linewidth": 1, "linestyle": "-"},
)



plt.show()
```
